```python
from pathlib import Path
from PIL import Image
import torch
from transformers import BlipProcessor, BlipForConditionalGeneration
import pyttsx3

# --- 1. Load BLIP model (image captioning) ---
device = "cuda" if torch.cuda.is_available() else "cpu"
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base").to(device)

# --- 2. Define function to generate caption ---
def caption_image(img_path: str) -> str:
    image = Image.open(img_path).convert("RGB")
    inputs = processor(images=image, return_tensors="pt").to(device)
    with torch.no_grad():
        out = model.generate(**inputs, max_new_tokens=30)
    return processor.decode(out[0], skip_special_tokens=True).strip()

# --- 3. Text-to-Speech (save as wav) ---
def say_to_wav(text: str, out_wav: str):
    engine = pyttsx3.init()
    engine.save_to_file(text, out_wav)
    engine.runAndWait()
    return out_wav

# --- 4. Run on your image ---
img_path = r"C:\Users\lgunupud\Downloads\image.png"  # <= raw string for Windows path
caption = caption_image(img_path)
print("Generated caption:", caption)

# Save audio
out_wav = str(Path(img_path).with_suffix(".wav"))
say_to_wav(caption, out_wav)
print(f"Audio saved at: {out_wav}")
```
